# Pré-Processamento de Dados

### Normalização de Dados

* **É definida como a raiz quadrada do somatório de todos os elementos do vetor ao quadrado**.

```python
# Carregando bibliotecas

import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import Normalizer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Gerando dados de exemplos

dados = [[-1, 2], [0.5, 6], [0, 10], [1, 18]]
dados
```

```python
# Verificando o valor da média

np.mean(dados)
```

```python
# Visualizando a distribuição dos dados

sns.distplot(np.array(dados).reshape(-1))
plt.show()
```

**Instânciando o objeto para normalizar os dados**

```python
# Instânciando o objeto

normal = Normalizer().fit(dados)
```

```python
# Aplicando a normalização nos dados, atribuindo a nova variável e alterando dados para 1 dimensão

dados_2 = normal.transform(dados) # Normalizando os dados
dados_2 = dados_2.reshape(-1)     # Alterando a dimensão
```

```python
# Verificando a média

np.mean(dados_2)
```

```python
# Visualizando a distribuição dos dados

sns.distplot(dados_2)
plt.show()
```
