# Pré-Processamento de Dados

**StandardScaler**

* Padroniza a escala dos dados sem inferir na sua forma;
* Mantém a escala dos dados mais fiel que a Técnica MinMaxScale.

```python
# Carregando bibliotecas

import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import Normalizer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Simulando um conjunto de dados com distribuição normal

dados = np.random.normal(size = 50)
```

```python
# Alterando a forma dos dados para 1 dimensão

dados = dados.reshape(-1, 1)
```

```python
# Visualizando distribuição dos dados

sns.distplot(dados)
plt.show()
```

```python
# Carregando bibliotecas

from sklearn.preprocessing import StandardScaler
```

```python
# Instânciando o objeto

modelo = StandardScaler().fit(dados)
```

```python
# Transformando os dados e alterando para apenas 1 dimensão

dados_2 = modelo.transform(dados)
dados_2 = dados_2.reshape(-1, 1)
```

```python
# Visualizando distribuição dos dados

sns.distplot(dados_2)
plt.show()
```
