# Pré-Processamento de Dados

* **Standartization**

```python
# Carregando bibliotecas

import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import Normalizer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Simulando um conjunto de dados com distribuição normal

dados = np.random.normal(size = 50)
```

```python
# Alterando a forma dos dados para 1 dimensão

dados = dados.reshape(-1, 1)
```

```python
# Visualizando distribuição dos dados

sns.distplot(dados)
plt.show()
```

```python
# Carregando biblioteca

from sklearn.preprocessing import MinMaxScaler
```

```python
# Instânciando o objeto

modelo = MinMaxScaler().fit(dados)
```

```python
# Tranformando os dados e alterando sua dimensão

dados_2 = modelo.transform(dados) # Transformando os dados
dados_2 = dados_2.reshape(-1, 1)  # Alterando a sua dimensão
```

```python
# Visualizando a distribuição dos dados

sns.distplot(dados_2)
plt.show()
```
