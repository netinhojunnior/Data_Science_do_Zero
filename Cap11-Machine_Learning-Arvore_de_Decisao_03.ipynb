```python
# Carregando bibliotecas

import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
# Carregando base de dados

dados = pd.read_csv('/content/drive/MyDrive/Meus Scripts/Data Science do Zero/Dados/bank-numeric.csv')
```

```python
# Visualizando base de dados

dados.head(3)
```

**Separando os dados em previsores e classes**

```python
previsores = dados.drop('deposit_cat', axis = 1)
classes = dados.deposit_cat
```

**Separando os dados em treino e teste: 70% para treino e 30% para teste**

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(previsores, classes, test_size = 0.3)
```

**Treinando o modelo**

```python
# Criando uma função para treinar o modelo e comparar o parâmetro max_depth

def comparando_modelo(MaxDepth):
  if MaxDepth == 0:
    model = tree.DecisionTreeClassifier(random_state = 1)
  else:
    model = tree.DecisionTreeClassifier(random_state = 1, max_depth = MaxDepth)
  model.fit(X_treino, y_treino)
  score_treino = model.score(X_treino, y_treino)
  score_teste = model.score(X_teste, y_teste)
  return score_treino, score_teste
```

```python
# Testando a função

comparando_modelo(2)
```

```python
# Exibindo resultados

print('{:10} {:20} {:20}'.format('depth', 'Score de Treino', 'Score de Teste'))
print('{:10} {:20} {:20}'.format('-----', '---------------', '--------------'))
print('{:1}         {} '.format(2, str(comparando_modelo(2))))
print('{:1}         {} '.format(3, str(comparando_modelo(3))))
print('{:1}         {} '.format(4, str(comparando_modelo(4))))
print('{:1}         {} '.format(10, str(comparando_modelo(10))))
print('{:1}         {} '.format(15, str(comparando_modelo(15))))
print('{:1}         {} '.format('Full', str(comparando_modelo(0))))
```
